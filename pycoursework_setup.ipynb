## Overview of Data

### Checking sample size in each year

```python
dfA = pd.read_sql_query('''SELECT Year, COUNT(FlightNum) AS total_flight 
                        FROM 'Y05_06'
                        GROUP BY Year''',conn)
dfA
```

Includes all observations. Even those with missing values. 

### Overview of Delayed Flights
Filtering out Cancelled and Diverted flights

```python
dfB = pd.read_sql_query('''SELECT Year, 
                COUNT(FlightNum) AS flights,
                AVG(ArrDelay) AS avg_arr_delay,
                MAX(ArrDelay) AS max_arr_delay,
                MIN(ArrDelay) AS min_arr_delay
                FROM 'Y05_06'
                WHERE Cancelled = 0 AND Diverted = 0
                GROUP BY Year''',conn)
dfB
```

```python
C = pd.merge(
    dfA,
    dfB,
    on='Year')
C
```

Saving C as a csv file

```python
C.to_csv('pysetupdata.csv',header = True, index = False)
```

### Checking if the dataset consist of only the USA

We filter out all the destination, dest, listed in Y05_06. To cross check with the list of airports. This can be done by creating a data frame 'check' and then parsing in our dest and inner join with the airports.

```python
dest = pd.read_sql_query("SELECT Dest From Y05_06", conn)
dest
```

```python
airports = pd.read_sql_query("SELECT * From airports",conn)
airports
```

```python
check = pd.merge(dest,airports,how = "inner", left_on = "Dest", right_on = "iata")
```
